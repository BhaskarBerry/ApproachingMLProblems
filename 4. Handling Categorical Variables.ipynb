```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import os
warnings.filterwarnings('ignore')
```

```python
path = os.path.abspath("Data/cat_in_dat_train.csv")
```

```python
data = pd.read_csv(path)
data.head()
```

```python
plt.figure(figsize=(10,5))
x=data.target.value_counts()
plt.bar(x.index,x)
plt.gca().set_xticks([0,1])
plt.title('Distribution of target variable')
plt.show()
```

```python
mapping = {
    'Freezing': 0,
    'Warm': 1,
    'Cold':2, 
    'Boiling Hot':3,
    'Hot':4, 
    'Lava Hot':5
    }
```

```python
# Label Encoding
data.loc[:,"ord_2"] = data.ord_2.map(mapping)
```
